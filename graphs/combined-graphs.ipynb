```python
import matplotlib.pyplot as plt
import yaml
from datetime import datetime, timedelta
import matplotlib.dates as mdates

with open('../covid-19.yaml') as file:
    dataFile = yaml.load(file, Loader=yaml.FullLoader)
    worldData = dataFile['World']
    countryData = dataFile['Country']
```

```python
statData = []
for entry in worldData:
    statData.append([entry["Date"], entry["Confirmed"], entry["Deaths"]])
statData.sort()
x, y, z = zip(*statData)

plt.figure(figsize=(12,7))
plt.plot(x, y, label="Cases")
plt.plot(x, z, label="Deaths")
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.title("World Statistics")
plt.xlabel("Dates")
plt.legend()
plt.show()
```
