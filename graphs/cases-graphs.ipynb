```python
import matplotlib.pyplot as plt
import yaml
from datetime import datetime, timedelta
import matplotlib.dates as mdates

with open('../covid-19.yaml') as file:
    dataFile = yaml.load(file, Loader=yaml.FullLoader)
    worldData = dataFile['World']
    countryData = dataFile['Country']
```

```python
print("Current Country Cases Comparrison:")
print("-----------------------")
statData = []
for country in countryData:
    statData.append([country, countryData[country][0]["Confirmed"]])
statData.sort(key=lambda x:x[1], reverse=True)
for dataPiece in statData:
    print("{:<15s}{:<7}".format(dataPiece[0], dataPiece[1]))
```

```python
statData = []
for entry in worldData:
    statData.append([entry["Date"], entry["Confirmed"]])
statData.sort()
x, y = zip(*statData)

plt.figure(figsize=(12,7))
plt.plot(x, y)
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.title("World Cases")
plt.xlabel("Dates")
plt.ylabel("Cases")
plt.show()
```

```python
plt.figure(figsize=(12,7))
for country in countryData:
    statData = []
    for entry in countryData[country]:
        statData.append([entry["Date"], entry["Confirmed"]])
    statData.sort()
    x, y = zip(*statData)
    plt.plot(x,y,label=country)

plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.title("Combined Country Cases")
plt.xlabel("Dates")
plt.ylabel("Cases")
plt.legend()
plt.show()
```

```python
statData = []
for entry in countryData["UK"]:
    statData.append([entry["Date"], entry["Confirmed"]])
statData.sort()
x, y = zip(*statData)

plt.figure(figsize=(12,7))
plt.plot(x, y)
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.title("United Kingdom Cases")
plt.xlabel("Dates")
plt.ylabel("Cases")
plt.show()
```

```python
statData = []
size = len(countryData["UK"])
for i in range(size):
    date = countryData["UK"][i]["Date"]
    a = countryData["UK"][i]["Confirmed"]
    b = a
    if(i+1 < size):
        b = countryData["UK"][i+1]["Confirmed"]
    statData.append([date, a-b])
statData.sort()
x, y = zip(*statData)

plt.figure(figsize=(12,7))
plt.plot(x, y)
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.title("United Kingdom Cases Growth")
plt.xlabel("Dates")
plt.ylabel("Cases")
plt.show()
```
