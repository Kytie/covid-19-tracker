```python
import matplotlib.pyplot as plt
import yaml
from datetime import datetime, timedelta
import matplotlib.dates as mdates

with open('../covid-19.yaml') as file:
    dataFile = yaml.load(file, Loader=yaml.FullLoader)
    worldData = dataFile['World']
    countryData = dataFile['Country']
```

```python
statData = []
for entry in worldData:
    statData.append([entry["Date"], entry["Deaths"]])
statData.sort()
print()
x, y = zip(*statData)
plt.figure(figsize=(12,7))
plt.plot(x, y)
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.title("World Deaths")
plt.xlabel("Dates")
plt.ylabel("Deaths")
plt.show()
```

```python
plt.figure(figsize=(12,7))
for country in countryData:
    statData = []
    for entry in countryData[country]:
        statData.append([entry["Date"], entry["Deaths"]])
    statData.sort()
    x, y = zip(*statData)
    plt.plot(x,y,label=country)

plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.title("Combined Country Deaths")
plt.xlabel("Dates")
plt.ylabel("Deaths")
plt.legend()
plt.show()
```
